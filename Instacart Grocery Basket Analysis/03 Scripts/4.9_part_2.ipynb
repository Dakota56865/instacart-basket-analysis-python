# Importing Libraries

```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

## Importing Data

```python
#create path
path = r'C:\Users\smerg\Instacart Grocery Basket Analysis'
# imporing new merged data with customers
instacart = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared data', 'ords_prods_cust_merged_data.pkl'))
```

### Task 2

```python
instacart.head()
```

```python
# creating Bar Chart
bar_2 = loyalty_counts = instacart['loyalty_flag'].value_counts()
loyalty_counts.plot.bar(color='skyblue')
plt.title('Distribution of Loyalty Flags')
plt.xlabel('Loyalty Flag')
plt.ylabel('Count')
plt.show()
```

```python
# creating Histogram
hist_3 = instacart['order_hour_of_day'].plot.hist(bins=24, color='lightgreen')
plt.title('Distribution of Orders by Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Frequency')
plt.show()
```

```python
# creating a Scatterplot
scat = sns.scatterplot(data=instacart, x='age', y='income')
plt.title('Scatterplot of Age vs. Income')
plt.xlabel('Age')
plt.ylabel('Income')
plt.show()
```

```python
# creating a Line Chart
line_3 = sample_data = instacart.sample(n=1000, random_state=42)
sample_data.groupby('order_hour_of_day')['prices'].mean().plot.line(color='purple')
plt.title('Average Order Prices by Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Average Price')
plt.show()
```
