```python
# Renaming columns for consistency
customer_df.rename(columns={
    'user_id': 'user_id',
    'First Name': 'first_name',
    'Surnam': 'surname',
    'Gender': 'gender',
    'STATE': 'state',
    'Age': 'age',
    'date_joined': 'date_joined',
    'n_dependants': 'number_of_dependants',
    'fam_status': 'family_status',
    'income': 'income'
}, inplace=True)

# Check for any unnecessary columns (for this task, assume all columns are needed)
# Dropping any unnecessary columns (none in this case)
# customer_df.drop(columns=['unnecessary_column'], inplace=True)

# Display the updated dataframe
customer_df.head()
```

```python
# Check for missing values
print(customer_df.isnull().sum())

# Handling missing values (if any)
# For example, you can fill missing values or drop rows/columns
customer_df.dropna(inplace=True)

# Check for duplicates
duplicates = customer_df.duplicated().sum()
print(f'Duplicates: {duplicates}')

# Remove duplicates if necessary
customer_df.drop_duplicates(inplace=True)

# Check for mixed types
for col in customer_df.columns:
    print(f'{col}: {customer_df[col].dtype}')

# Convert any mixed-type data if necessary (e.g., converting columns to correct types)
customer_df['age'] = customer_df['age'].astype(int)

# Display cleaned dataframe
customer_df.head()
```

```python
# Ensure user_id is of the same data type in both DataFrames
customer_df['user_id'] = customer_df['user_id'].astype('int64')
ords_prods_merge['user_id'] = ords_prods_merge['user_id'].astype('int64')

# Merge the dataframes on 'user_id'
ords_prods_cust_merged_data = pd.merge(ords_prods_merge, customer_df, on='user_id', how='inner')

# Check the merged DataFrame
ords_prods_cust_merged_data.head()
```
