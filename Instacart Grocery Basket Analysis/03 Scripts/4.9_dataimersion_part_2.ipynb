## Part 2

```python
# imporing new merged data with customers
instacart = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared data', 'ords_prods_cust_merged_data.pkl'))
```

```python
# Histogram of order_hour_of_day
hist_3 = instacart['order_hour_of_day'].plot.hist(bins=24, color='lightblue')
plt.title('Distribution of Orders by Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Frequency')
plt.show()
```

### Analysis of Order Distribution by Hour of the Day
The histogram above illustrates the frequency of orders placed at different hours of the day. It shows that most orders are concentrated around mid-morning to early afternoon, with a peak around 10 AM. This indicates that customers are most active during these hours, which could be crucial for targeting promotions.


```python
import os
import matplotlib.pyplot as plt

# Bar Chart of loyalty_counts
loyalty_counts = instacart['loyalty_flag'].value_counts()
ax = loyalty_counts.plot.bar(color='orange')  # Get the axes object

# Add titles and labels
plt.title('Distribution of Customer Loyalty')
plt.xlabel('Loyalty Flag')
plt.ylabel('Number of Customers')

# Ensure the directory exists
save_path = os.path.join(path, '04 Analysis', 'Visualizations')
os.makedirs(save_path, exist_ok=True)

# Save the plot
fig = ax.get_figure()  # Get the figure from the axes object
fig.savefig(os.path.join(save_path, 'bar_3_loyalty_flag.png'))

# Display the plot
plt.show()

# Close the plot
plt.close(fig)
```

```python
# Bar Chart of loyalty_counts
loyalty_counts = instacart['loyalty_flag'].value_counts()
loyalty_counts.plot.bar(color='orange')
plt.title('Distribution of Customer Loyalty')
plt.xlabel('Loyalty Flag')
plt.ylabel('Number of Customers')
plt.show()
```

```python
# Line Chart for Prices by Hour of Day
line_3 = instacart.groupby('order_hour_of_day')['prices'].mean().plot.line(color='green')
plt.title('Average Expenditure by Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Average Prices')
plt.show()
```

```python
# Line Chart for Age vs. Number of Dependents
line_4 = instacart.groupby('age')['number_of_dependants'].mean().plot.line(color='blue')
plt.title('Average Number of Dependents by Age')
plt.xlabel('Age')
plt.ylabel('Average Number of Dependents')
plt.show()
```

### Analysis of Age and Family Situation
The line chart shows the average number of dependents by age. We observe that the number of dependents generally increases with age up to a certain point and then begins to decrease as customers reach their later years. This trend may reflect typical family life cycles, with younger adults having fewer dependents and middle-aged adults having the most.


```python
# Scatterplot of Age vs. Income
scat = sns.scatterplot(data=instacart, x='age', y='income', color='red')
plt.title('Scatterplot of Age vs. Income')
plt.xlabel('Age')
plt.ylabel('Income')
plt.show()
```

### Analysis of Age and Spending Power
The scatterplot above illustrates the relationship between age and income. It shows that income generally increases with age, peaking around middle age, before gradually decreasing in older age. This pattern likely reflects career progression, with individuals earning more as they gain experience and then reducing their income as they approach retirement.
