#### function

```python
# define a function
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# applying the function
df['price_range'] = df.apply(price_label, axis = 1)
```

```python
# Create a new column 'busiest_day' based on the day of the week with the most orders
busiest_day = ords_prods_merge['order_dow'].mode()[0]
ords_prods_merge['busiest_day'] = ords_prods_merge['order_dow'].apply(
    lambda x: 'Busiest day' if x == busiest_day else 'Not busiest day'
)
```

```python
# Step 1: Identify the two busiest and two slowest days
day_counts = ords_prods_merge['order_dow'].value_counts()

# The two busiest days
busiest_days = day_counts.nlargest(2).index.tolist()

# The two slowest days
slowest_days = day_counts.nsmallest(2).index.tolist()

# Step 2: Create a new column 'busiest_days'
def label_busiest_days(day):
    if day in busiest_days:
        return 'Busiest days'
    elif day in slowest_days:
        return 'Slowest days'
    else:
        return 'Regular days'

ords_prods_merge['busiest_days'] = ords_prods_merge['order_dow'].apply(label_busiest_days)
```

```python
# Check the value counts for the 'busiest_days' column
print(ords_prods_merge['busiest_days'].value_counts())
```

```python
result = []

for value in ords_prods_merge["order_dow"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
#Combining the result to the dataframe to give it meaning. Creating the new summary column
ords_prods_merge['busiest_day'] = result
```

```python
# Step 1: Find the two busiest and two slowest days
day_counts = ords_prods_merge['order_hour_of_day'].value_counts()

# The two busiest days
busiest_days = day_counts.nlargest(2).index.tolist()

# The two slowest days
slowest_days = day_counts.nsmallest(2).index.tolist()

# Step 2: Create a new column 'busiest_days'
def label_busiest_days(day):
    if day in busiest_days:
        return 'Busiest days'
    elif day in slowest_days:
        return 'Slowest days'
    else:
        return 'Regular days'

ords_prods_merge['busiest_days'] = ords_prods_merge['order_hour_of_day'].apply(label_busiest_days)
```

```python
# Check the value counts for the 'busiest_days' column
print(ords_prods_merge['busiest_days'].value_counts())
```

```python
# Step 1: Calculate the frequency of orders by hour
hourly_orders = ords_prods_merge['order_hour_of_day'].value_counts()

# Define thresholds for busiest, average, and slowest periods
most_orders = hourly_orders.nlargest(4).index.tolist()  # For example, top 4 hours
fewest_orders = hourly_orders.nsmallest(4).index.tolist()  # For example, bottom 4 hours

# Step 2: Create a new column 'busiest_period_of_day'
def label_busiest_period(hour):
    if hour in most_orders:
        return 'Most orders'
    elif hour in fewest_orders:
        return 'Fewest orders'
    else:
        return 'Average orders'

ords_prods_merge['busiest_period_of_day'] = ords_prods_merge['order_hour_of_day'].apply(label_busiest_period)
```

```python
# Print the frequency of the new 'busiest_period_of_day' column
print(ords_prods_merge['busiest_period_of_day'].value_counts())
```
